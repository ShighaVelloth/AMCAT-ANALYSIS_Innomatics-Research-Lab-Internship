```python
#checking for outliers in numerical columns
#Using Interquartile Range (IQR)


numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Select only numeric columns
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Filtering for outliers
outlier_mask = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))

# Check if any outliers exist
if outlier_mask.any().any(): 
    print("Yes")
else:
    print("No")
```

## 4 STATISTICAL TESTING

## 4.1 Research Question 1
Times of India article (dated Jan 18, 2019) states: “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer, or Associate Engineer, you can earn up to 2.5-3 lakhs as a fresh graduate.”

Goal: We need to statistically test this claim using the given dataset.

## Hypothesis Testing

## T- test

Null Hypothesis (H₀): The average salary for fresh graduates in roles such as Programming Analyst, Software Engineer, Hardware Engineer, or Associate Engineer after completing Computer Science Engineering is ₹275,000.

H₀:μ=275,000

Alternative Hypothesis (H₁): The average salary for fresh graduates in these roles is not ₹275,000

H₁ :μ!=275,000

This is a one-tailed t-test where we test if the mean salary is ₹275,000.

```python
import scipy.stats as stats

# Step 1: Filter the dataset for Computer Science graduates in specific roles
roles = ['programmer analyst', 'software engineer', 'system engineer', 'trainee engineer']
filtered_df = df[(df['Specialization'] == 'computer science & engineering') & (df['Designation'].isin(roles))]

# Step 2: Check the average salary for the filtered dataset
mean_salary = filtered_df['Salary'].mean()
```

```python
# Step 3: Perform two-sample t-tests

threshold = 275000

# T-test for lower bound (Salary > ₹2.5 lakhs)
t_stat, p_value = stats.ttest_1samp(filtered_df['Salary'], threshold)

```

```python
print(t_stat)
```

```python
print(p_value)
```

```python
# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. The average salary is significantly different from ₹275,000.")
else:
    print("Fail to reject the null hypothesis. The average salary is not significantly different from ₹275,000.")
```

## 4.2 Research Question 2

##  Does the preference of Specialisation depend on the Gender

## Chi- Square test

Null Hypothesis (H₀): The preference of Specialization does not depend on Gender (they are independent).

Alternative Hypothesis (H₁): The preference of Specialization depends on Gender (they are not independent).

```python
from scipy.stats import chi2_contingency

# Create a contingency table of Gender and Specialization
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Perform the Chi-Square test of independence
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

# Print the Chi-Square statistic, p-value, and degrees of freedom
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_val}")
print(f"Degrees of Freedom: {dof}")
```

```python
# Interpret the results
if p_val < 0.05:
    print("Reject the null hypothesis: The preference of Specialization depends on Gender.")
else:
    print("Fail to reject the null hypothesis: The preference of Specialization does not depend on Gender.")
```

## 4.3 Research Question 3 

## Is there a significant relationship between gender and salary

Null Hypothesis (H₀): There is no significant difference in the average salary between male and female.
    
Alternative Hypothesis (H₁): There is a significant difference in the average salary between male and female.

```python
male_salaries = df[df['Gender'] == 'm']['Salary']
female_salaries = df[df['Gender'] == 'f']['Salary']

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(male_salaries, female_salaries, nan_policy='omit')

# Output the result
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
# Interpret the result
if p_val < 0.05:
    print("Reject the null hypothesis: There is a significant difference in salary based on gender.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in salary based on gender.")
```
